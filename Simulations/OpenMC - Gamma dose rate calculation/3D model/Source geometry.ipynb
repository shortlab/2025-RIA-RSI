```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Source_height = 21.11 #cm
Source_pitchcirclediameter = 20.91 #cm
Nb_pencils=48
Diam_pencil=np.pi*Source_pitchcirclediameter/Nb_pencils

def plotcircle(r,origin=(0.,0.),color=None):
    theta=np.linspace(0,2*np.pi,1000)
    return plt.plot(r*np.cos(theta)+origin[0],r*np.sin(theta)+origin[1],color=color,linewidth=0.1)

plt.figure()
plotcircle(Source_pitchcirclediameter/2,color='black')
for theta in [2*np.pi*i/Nb_pencils for i in range(Nb_pencils)]:
    plotcircle(Diam_pencil/2,origin=(Source_pitchcirclediameter/2*np.cos(theta),Source_pitchcirclediameter/2*np.sin(theta)),color='blue')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.xlim(left=9.5,right=11.5)
plt.ylim(bottom=-1,top=1)
plt.show()
```
